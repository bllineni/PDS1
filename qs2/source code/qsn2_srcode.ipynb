```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\user\\Desktop\\StudentsPerformance.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
df['lunch'] = pd.Categorical(df['lunch'])
df['test preparation course'] = pd.Categorical(df['test preparation course'])
```

```python
df[['math score', 'reading score', 'writing score']] = df[['math score', 'reading score', 'writing score']].astype(int)
```

```python
df.to_csv("C:/Users/user/Documents/pds/qs2/Clean data/cln_data.csv", index=False)
```

```python
print(df.head())
```

```python
gender_counts = df["gender"].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.show()
```

```python

test_scores = df[["math score", "reading score", "writing score"]].mean()
plt.plot(test_scores.index, test_scores.values)
plt.title("Test Scores Over Time")
plt.xlabel("Test")
plt.ylabel("Average Score")
plt.xticks(range(3), ["Math", "Reading", "Writing"])
plt.show()
```

```python
pivot_table = df.pivot_table(index="parental level of education", columns="gender", values=["math score", "reading score", "writing score"], aggfunc="mean")
sns.heatmap(pivot_table, cmap="YlGnBu")
plt.title("Exam Scores by Gender and Education Level")
plt.xlabel("Gender")
plt.ylabel("Education Level")
plt.show()
```

```python
sns.scatterplot(x="math score", y="reading score", data=df)
plt.title("Math Scores vs. Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
sns.boxplot(x="parental level of education", y="math score", data=df)
sns.boxplot(x="parental level of education", y="reading score", data=df)
sns.boxplot(x="parental level of education", y="writing score", data=df)
plt.title("Exam Scores by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.hist(df['math score'], bins=20)
plt.xlabel('Math Scores')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.catplot(x='test preparation course', hue='lunch', kind='count', data=df)
plt.show()
```

```python
prep_course = df.groupby('gender')['test preparation course'].value_counts().unstack()
prep_course.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Test Preparation Course Completion by Gender')
plt.show()
```
